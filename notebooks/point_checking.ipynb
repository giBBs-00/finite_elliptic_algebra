```python
from elliptic_algebra import EllipticCurve, GF
from typing import Tuple
```

```python

from pprint import pprint

def opt_pt(p, a, c = 0):
    curves = []
    gf = GF(p, a)
    elements = gf.elements

    if not gf.is_primitive:
        raise "Not primitive"

    for x in gf.elements:
        if x == gf([0]):
            continue
        for y in elements:
            if y == gf([0]):
                continue
            P = EllipticCurve.FreeECPoint(x, y)
            curves = P.singular_curves()
            n = len(curves)
            if n == 1:
                if c == 0:
                    return (gf, P)
                else:
                    c = c - 1



def split(input: Tuple[GF, EllipticCurve.FreeECPoint]):
    gf, pt = input
    x = pt.x
    y = pt.y


    return [EllipticCurve.FreeECPoint(x * gf.a(i * (((gf.p ** gf.n) - 1) // 3)), y * gf.a(j * (((gf.p ** gf.n) - 1) // 2))) for j in [0,1] for i in [0,1,2]]


split_check = [(x.x, x.y) for x in split(opt_pt(5, [1, 1, 2], 17))]
print(split_check)
interest = [{"point": (x.x, x.y), "singular_curves": x.candidate_curves(), "count": len(x.candidate_curves())} for x in split(opt_pt(5, [1, 1, 2], 17))]


all_curves = []
for elem in interest:
    singular_curves = elem["singular_curves"]
    for curve in singular_curves:
        all_curves.append(f"{curve}")
    
# Sort the curves based on their string representation
sorted_curves = sorted(all_curves)


print("All curves:")
for curve in sorted_curves:
    print(curve)
print(len(set(sorted_curves)))


```
